```python
import okex.futures_api as future
import time
api_key = ''
seceret_key = ''
passphrase = ''

while True:
    futureAPI = future.FutureAPI(api_key, seceret_key, passphrase, True)
    result = futureAPI.get_specific_ticker('BTC-USD-191227')
    print (result['best_bid'],result['timestamp']) 
    time.sleep( 1 )
```
